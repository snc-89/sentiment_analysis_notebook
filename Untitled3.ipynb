```python
foo = pd.read_csv('files/concat_df_sans_zero.csv')
```

```python
bin_labels = [0, 1, 2, 3, 4, 5]
foo['ratio_bins'] = pd.qcut(foo['ratio'], q=6, labels = bin_labels)
```

```python
_p = sns.lmplot("sentMean", "ratio", data=foo[foo.commentCount > 100], fit_reg=False, hue='ratio_bins', height=10)
```

```python
df = foo.drop(['Unnamed: 0','date','title','ratio','ratio_bins'],axis=1)
#idx = [0,1,2,3,4,7,8,9,11,14,15,16]
#df = df.iloc[:,idx]

df.columns
```

```python
X = df
y = foo['ratio_bins']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20,random_state=42)

models = [GaussianNB(),MLPClassifier(),KNeighborsClassifier(n_neighbors=7),DecisionTreeClassifier(),SVC()]
```

```python
for model in models:
    model = model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    
    print(accuracy_score(y_test,y_pred))
    #print(f1_score(y_test,y_pred,average='macro'))
    #print(precision_score(y_test,y_pred,average='macro'))
    #print(recall_score(y_test,y_pred,average='macro'))
    #print(confusion_matrix(y_test,y_pred))
```
