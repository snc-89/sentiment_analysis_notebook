```python
def get_csv_list(file):
    with open(file, encoding='utf-8') as f:
        r = csv.reader(f)
        return_list = list(r)
    return_list = return_list[0]
    f.close()
    return return_list
```

```python
classes = get_csv_list('classeslist2.csv')
for i in range(len(classes)):
    classes[i] = int(classes[i])
classes = classes + [1]*134
```

```python
vids['class'] = classes
```

```python
vids.groupby(['class']).count()
```

```python
view = vids[(vids['commentCount'] > 100) & (vids['class'] > -1) ]
plt.figure(figsize=(20, 20))

sns.scatterplot(data=view, x='sentiment',y='ratio',hue='class',palette=['y','r','g','b','Black','Brown','Orange'])
```
